# Consulta de terminos python

- **tic:** Originalmente de *MATLAB*, es un comando para comenzar un temporizador durante el tiempo que se ejecuta un código

- **toc:** Originalmente de *MATLAB*, es un comando para terminar el temporizador y hacer la lectura del tiempo comenzado previamente en la ejecución del código

En python, el equivalente a esto seria el uso de la libreria *time* que implementa tiempo, para allí podemos hacer una diferencia del tiempo de inicio y tiempo de finalización para determinar el tiempo de ejecución del código.

- **time:** Este modulo esta encargado en dar funciones para el manejo de cuestiones relacionadas con el tiempo.

- **tf:** Para python control las siglas tf indican la conversión de un array que se trabaja como polinomio a una función de transferencia (numerador, denominador).

- **ss:** Para python control las siglas ss indican la conversión de distintos arrays matriciales (A, B, C, D) a volverse un espacio de estado (SS) donde relacionan estado (x), entrada (u) y salida (y) dada unas ecuaciones de forma Ax + B

- **zpk:** Comando utilizado para crear una función de transferencia desde sus polos, ceros y su termino de ganancia

- **tf2ss:** Comando usado para el cambio de función de transferencia a espacio de estado

- **tf2zpk** Para python scipy, comando usado para el paso de función de transferencia a sus polos, ceros y factor de ganancia

- **ss2zp:** Para python scipy, comando usado para el paso de espacio de estados a sus polos, ceros y factor de ganancia

- **ss2tf:** Para python control, comando utilizado para el paso de espacio de estado a función de transferencia equivalente.

- **zp2ss:** Comando utilizado para convertir una representación de un sistema definida por **ceros, polos y ganancia** (ZPK) a su representación en **espacio de estado**. El resultado son las matrices \(A\), \(B\), \(C\) y \(D\) que describen la dinámica del sistema.

- **zp2tf:** Comando utilizado para convertir una representación de un sistema dada por **ceros, polos y ganancia** a su **función de transferencia equivalente**, obteniendo los polinomios de numerador y denominador.

- **roots:** Función utilizada para calcular las **raíces de un polinomio**. En control de sistemas se usa comúnmente para encontrar los **polos o ceros** de una función de transferencia a partir de sus coeficientes.

- **pole:** Comando utilizado para calcular los **polos de un sistema dinámico**, es decir, los valores de \(s\) que hacen que el denominador de la función de transferencia sea cero.

- **zero:** Comando utilizado para calcular los **ceros de un sistema**, es decir, los valores de \(s\) que hacen que el numerador de la función de transferencia sea cero.

- **poly:** Función utilizada para construir el **polinomio característico** a partir de un conjunto de raíces o valores propios. También puede utilizarse para generar los coeficientes de un polinomio a partir de sus raíces.

- **eig:** Función utilizada para calcular los **valores propios (eigenvalues)** de una matriz. En sistemas dinámicos, estos corresponden a los **polos del sistema cuando se trabaja en espacio de estado**.

- **pzmap:** Comando utilizado para graficar el **mapa de polos y ceros** de un sistema en el plano complejo \(s\). Permite analizar visualmente la estabilidad y comportamiento dinámico del sistema.

- **dcgain:** Comando utilizado para calcular la **ganancia en corriente continua (DC gain)** de un sistema, es decir, la ganancia cuando la frecuencia tiende a cero. Representa el valor final de la salida ante una entrada constante.

- **minreal:** Comando utilizado para obtener una **realización mínima del sistema**, eliminando polos y ceros que se cancelan entre sí para simplificar la representación matemática.

- **subplot:** Función utilizada para dividir una ventana de gráficos en **varias subgráficas**, permitiendo mostrar múltiples resultados dentro de la misma figura.

- **feedback:** Comando utilizado para crear un **sistema en retroalimentación (lazo cerrado)** a partir de un sistema en lazo abierto. Generalmente se usa para modelar sistemas de control con retroalimentación unitaria.

### Preguntas

- ¿Qué es el polinomio característico de un sistema?

Un polinomio característico se define como el polinomio que se obtiene al calcular el determinante de la diferencia entre un múltiplo escalar de la matriz identidad y una matriz cuadrada dada. Las raíces de este polinomio representan los valores característicos de la matriz.

- ¿Qué son los valores propios de un sistema?

Los valores propios describen la estabilidad de un sistema y suelen asociarse con el álgebra lineal. Una forma de comprenderlos es mediante su aplicación en la descripción de sistemas de ingeniería, como el sistema masa-resorte-amortiguador. 

- ¿Qué son el tiempo de estabilización y el sobre nivel porcentual de la respuesta temporal de un sistema ante
entrada tipo escalón?

**Tiempo de estabilización:** Es el tiempo que tarda la salida de un sistema en permanecer dentro de una banda cercana (generalmente ±2% o ±5%) al valor final después de aplicarse una entrada escalón.

**Sobre nivel porcentual:** Es el porcentaje en que la respuesta del sistema supera el valor final durante su respuesta transitoria ante una entrada escalón.



# ***(5.1)*** Modelo sin tiempo muerto

**Modelo de ft, sin tiempo muerto:**

$$
G(s) = \frac{K}{(\tau_{1} s+1)(\tau_{2} s+1)}
$$


Expandimos nuestra ecuación sin tiempo muerto para dejar las constantes claras, y luego partimos de esta forma a volverla una ecuación diferencial, gracias a que sabemos de antemano que $G(s) = \frac{Y(s)}{U(s)}$.

$$
G(s)=\frac{K}{T_1 T_2 s^2+\left(T_1+T_2\right) s+1}
$$

Por lo tanto:

$$
\left(T_1 T_2 s^2+\left(T_1+T_2\right) s+1\right) Y(s)=K U(s)
$$

Pasando al dominio del tiempo:

$$
T_1 T_2 y^{\prime \prime}(t)+\left(T_1+T_2\right) y^{\prime}(t)+y(t)=K u(t)
$$

Normalizar la ecuación diferencial
(piden que el coeficiente de la mayor derivada sea 1)
Dividimos todo por $T_1 T_2$ :

$$
y^{\prime \prime}(t)+\frac{T_1+T_2}{T_1 T_2} y^{\prime}(t)+\frac{1}{T_1 T_2} y(t)=\frac{K}{T_1 T_2} u(t)
$$

Definir Variables de Estado de Fase

$$
\begin{aligned}
& x_1(t)=y(t) \\
& x_2(t)=y^{\prime}(t)
\end{aligned}
$$


Entonces:

$$
x_1^{\prime}(t)=x_2(t)
$$


Ahora despejamos $y^{\prime \prime}(t)$ :

$$
y^{\prime \prime}(t)=-\frac{T_1+T_2}{T_1 T_2} y^{\prime}(t)-\frac{1}{T_1 T_2} y(t)+\frac{K}{T_1 T_2} u(t)
$$


Sustituyendo las variables de estado:

$$
x_2^{\prime}(t)=-\frac{1}{T_1 T_2} x_1-\frac{T_1+T_2}{T_1 T_2} x_2+\frac{K}{T_1 T_2} u
$$

Forma matricial:

$$
\begin{aligned}
\dot{x} & =A x+B u \\
y & =C x+D u
\end{aligned}
$$


Matriz A

$$
A=\left[\begin{array}{cc}
0 & 1 \\
-\frac{1}{T_1 T_2} & -\frac{T_1+T_2}{T_1 T_2}
\end{array}\right]
$$


Matriz B

$$
B=\left[\begin{array}{c}
0 \\
\frac{K}{T_1 T_2}
\end{array}\right]
$$


Matriz C

$$
C=\left[\begin{array}{ll}
1 & 0
\end{array}\right]
$$


Matriz D

$$
D=0
$$


> Concluyendo lo previamente realizado obtenemos nuestras respuesta y asignando los valores correspondientes


**Modelo de FT sin tiempo muerto:**


$$
\frac{1.2}{(59.7 s+1)(22.3 s+1)}
$$

**Modelo de EE sin tiempo muerto:**

$$
\begin{gathered}
A=\left[\begin{array}{cc}
0 & 1 \\
-0.00075114 & -0.061596
\end{array}\right] \\
B=\left[\begin{array}{c}
0 \\
0.0009015
\end{array}\right] \\
C=\left[\begin{array}{ll}
1 & 0
\end{array}\right] \\
D=0
\end{gathered}
$$

(donde $1 /\left(T_1 T_2\right)=1 / 1331.31$ y $K /\left(T_1 T_2\right)=1.2 / 1331.31$ ).

# ***(5.2)*** Aplicación en python

```python
import numpy as np
import control as cs
import matplotlib.pyplot as plt
import scipy as sp
```

```python
#Comprobación desde los valores dictados - se calculan polos y ceros y se normaliza el denominador
# num y den para funcion de transferencia
num = np.array([1.2])
den = np.array([1331.31, 82, 1])

# creacion de modelos (FT y EE)
Gptf = cs.tf(num, den)
Mpss = cs.tf2ss(num,den)

print("Función de transferencia:")
print(Gptf)

print("Modelo en espacio de estado:")
print(Mpss)

```

```python
Gptf_response = cs.step_response(0.23 *Gptf)
Mpss_response = cs.step_response(0.23 *Mpss)

plt.figure(figsize=(16,6))

# Gráfica izquierda
plt.subplot(1,2,1)
plt.plot(Gptf_response.time, Gptf_response.outputs, label='Función de Transferencia')
plt.title('Respuesta al Escalón - Función de Transferencia')
plt.xlabel('Tiempo (s)')
plt.ylabel('Salida')
plt.grid(True)
plt.legend()

# Gráfica derecha
plt.subplot(1,2,2)
plt.plot(Mpss_response.time, Mpss_response.outputs, label='Espacio de Estado', color='orange')
plt.title('Respuesta al Escalón - Espacio de Estado')
plt.xlabel('Tiempo (s)')
plt.ylabel('Salida')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

print(f"Función de Transferencia - Tiempo de Establecimiento: {cs.step_info(0.23 * Gptf)['SettlingTime']:.2f}s")
print(f"Espacio de Estado - Tiempo de Establecimiento: {cs.step_info(0.23 * Mpss)['SettlingTime']:.2f}s")
print(f'Funcion de Transferencia - Valor de estabilización: {cs.step_info(0.23 * Gptf)["SteadyStateValue"]:.2f}')
print(f'Espacio de Estado - Valor de estabilización: {cs.step_info(0.23 * Mpss)["SteadyStateValue"]:.2f}')
```

**Justificación:** Luego de haberlo comprobado con la función *step_info* y acceder a los valores que guarda para el funcionamiento de nuestro sistema, se puede analizar de forma mátematica al tener las expresiones concretadas para revisar como funcionaría nuestro sistema.

***(5.2.1)*** - Visualmente desde la gráfica, vemos como alrededor de los 300 ~ 400 segundos, nuestro sistema se comienza a establizar en 1.2, visto de forma mátematica al analizar $0.23 * G(0) = 0.23 * \frac{1.2}{1331.31(0)^2 + 82(0) + 1} = 0.23 * \frac{1.2}{1} = 0.23 * 1.2 = 0.28$ de allí tenemos nuestra respuesta de la salida ante una entrada constante que es un escalón $u(t)$


***(5.2.2)*** - Estimados por la visualización de la gráfica se podría decir que ronda los $300s$, no obstante al poder verificar desde *step_info*, vemos que el valor real (aparte de ser igual para ambos sistemas) es $261.59 s$

**Nota:** Al hacerlo por el multiplo $0.23$ que menciona la guia con *step(0.23 * Gptf)*, la respuesta quedaria escalada con respecto a la que dicta el sistema desde su ganancia, quiza este factor se utiliza para normalizar la tendencia de respuesta a un valor en concreto, en este caso casi a 0.3.

# ***(5.3)*** Modelo con tiempo muerto

***(5.3.1)*** y ***(5.3.2)***  Python y gráficas

```python
# Tiempo muerto
Td = 39.2

# Aproximación de Padé (orden 3 para mejor precisión)
num_delay, den_delay = cs.pade(Td, 3)

# Crear función de transferencia del delay
delay_tf = cs.tf(num_delay, den_delay)

# Crear versión en espacio de estado del delay
delay_ss = cs.ss(delay_tf)

# Agregar tiempo muerto a los modelos
Gptfm = Gptf * delay_tf # agregado el delay a la función de transferencia
Mpssm = cs.series(delay_ss, Mpss) # agregado el delay al espacio de estado desde series

# Vector de tiempo suficientemente largo para ver el delay
t = np.linspace(0, 200, 2000)

# Respuesta al escalón
t1, y1 = cs.step_response(Gptfm, T=t)
t2, y2 = cs.step_response(Mpssm, T=t)

# vector de tiempo suficientemente largo para ver el tiempo muerto
t = np.linspace(0, 450, 1000)

# respuesta al escalón de magnitud 0.23
t1, y1 = cs.step_response(0.23 * Gptfm, T=t)
t2, y2 = cs.step_response(0.23 * Mpssm, T=t)

# crear nueva figura
plt.figure(figsize=(10,8))

# subplot 1 (arriba) -> Gptfm
plt.subplot(2,1,1)
plt.plot(t1, y1, label='Función de Transferencia con Delay')
plt.title("Respuesta al escalón (0.23) - Gptfm")
plt.xlabel("Tiempo (s)")
plt.ylabel("Salida")
plt.grid()
plt.legend()

# subplot 2 (abajo) -> Mpssm
plt.subplot(2,1,2)
plt.plot(t2, y2, label='Espacio de Estado con Delay', color='orange')
plt.title("Respuesta al escalón (0.23) - Mpssm")
plt.xlabel("Tiempo (s)")
plt.ylabel("Salida")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()

```

Las estructuras obtenidas mediante conversión representan el mismo sistema dinámico que los modelos originales. # ***(5.5)*** Sistema entrada salida Gptfm

```python
# vector de tiempo (suficientemente largo por el tiempo muerto)
t = np.linspace(0, 450, 1000)

# respuesta al escalón unitario
t_resp, y_resp = cs.step_response(Gptfm, T=t)

# nueva ventana de gráfica
plt.figure(figsize=(14,6))
plt.plot(t_resp, y_resp)
plt.title("Respuesta y(t) ante escalón unitario u(t)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Salida y(t)")
plt.grid()
plt.tight_layout()
plt.show()
```

***(5.5.1)*** Expresiones que relacionan $Y(s)$ con $E(s)$ y $Y(s)$ con $P(s)$

**Funciones de transferencia del sistema**

De acuerdo con el diagrama de bloques, la entrada al proceso es:

$$
U(s) = M(s) + P(s)
$$

donde

$$
M(s) = G_c(s)E(s)
$$

y la salida del proceso está dada por

$$
Y(s) = G_{ptfm}(s)U(s)
$$

Sustituyendo:

$$
Y(s) = G_{ptfm}(s)\left(G_c(s)E(s) + P(s)\right)
$$

Distribuyendo:

$$
Y(s) = G_{ptfm}(s)G_c(s)E(s) + G_{ptfm}(s)P(s)
$$

**Relación entre Y(s) y E(s)**

$$
Y_r(s) = \frac{Y(s)}{E(s)} = G_c(s)\,G_{ptfm}(s)
$$

**Relación entre Y(s) y P(s)**

$$
Y_p(s) = \frac{Y(s)}{P(s)} = G_{ptfm}(s)
$$

***(5.5.2)*** Formación de $G_{\text{ctf}}$

```python
# obtener polos del sistema
polos = cs.poles(Gptfm)

# seleccionar polos complejos (parte imaginaria distinta de cero)
polos_complejos = [p for p in polos if abs(np.imag(p)) > 1e-6]

# tomar los dos polos complejos conjugados
z1, z2 = polos_complejos[:2]

# numerador del controlador
num = 0.3 * np.poly([z1, z2])

# denominador (polo con constante de tiempo igual al tiempo muerto)
den = [Td, 1]

# crear función de transferencia del controlador
Gctf = cs.tf(num, den)

print("Ceros usados:", z1, z2)
print(Gctf)
```

***(5.5.3)*** Formación de $G_{\text{atfm}}$

```python
# función de transferencia de lazo abierto
Gatfm = Gctf * Gptfm

# simplificar cancelando polos y ceros
Gatfm = cs.minreal(Gatfm)

print(Gatfm)
print(type(Gatfm))
```

La variable $G_{\text {atfm }}$ corresponde a la función de transferencia del sistema en lazo abierto obtenida al multiplicar el controlador $G_c(s)$ por la planta $G_{p t f m}(s)$.

Al cancelar polos y ceros (dicho en la documentación) esto indica que algunos de los comportamientos dinamicos se pierden por ser redundantes y deja la representación minima del sistema (dado por el termino *minreal*). 

Luego de haber hecho esta multiplicación y transformación del sistema, lo resultante sera nuevamente otra función de transferencia (siendo el caso de ingresar un SS, la función *minreal* retorna un SS).

***(5.5.4)*** Cambio de $E(s) = 0$ y $P(s) = U(t)$ (escalon unitario)

La salida se calcula con:

$$
Y_p(s)=G_{p t f m}(s) P(s)
$$

y como $P(s)$ es un escalón unitario, simplemente se debe obtener la respuesta al escalón de $G_{p t f m}(s)$.

```python
# vector de tiempo
t = np.linspace(0, 450, 1000)

# respuesta temporal yp(t)
t_resp, yp = cs.step_response(Gptfm, T=t)

# nueva ventana de figura
plt.figure(figsize=(14,6))

plt.plot(t_resp, yp)
plt.title("Respuesta yp(t) ante perturbación escalón unitario")
plt.xlabel("Tiempo (s)")
plt.ylabel("Salida yp(t)")
plt.grid()

plt.show()

# métricas de la respuesta
info = cs.step_info(yp, T=t_resp) 
print("Tiempo de subida:", info['RiseTime'])
print("Tiempo de asentamiento:", info['SettlingTime'])
print("Valor en estado estacionario:", info['SteadyStateValue'])
print("Sobrepaso:", info['Overshoot'])
```

**Similitudes**

- Ambas respuestas presentan el mismo tiempo muerto (~39.2 s) porque la planta 

- La forma de la respuesta es similar, ya que depende de los mismos polos del sistema.

- Ambas muestran un retardo inicial antes de comenzar a cambiar.

**Diferencias**

- En 5.5 la salida se debe a la entrada de referencia (error) que pasa por el controlador y la planta.

- En 5.5.4 la salida se debe a una perturbación directa en la entrada del proceso.

- La dependencia de la respuesta ya no esta dependiendo del controlador, solo depende de $G_{\text{ptfm}}(s)$

***(5.5.5)*** Cambio de $P(s) = 0$ y $E(s) = U(t)$ (escalon unitario)

La salida se calcula con:

$$
Y_r(s)=G_{c}(s) G_{\text{ptfm}}(s)
$$

este corresponde a la respuesta del sistema de $G_{\text{atfm}}(s)$

```python
# vector de tiempo
t = np.linspace(0,450,1000)

# respuesta yr(t)
t_resp, yr = cs.step_response(Gatfm, T=t)

# nueva figura
plt.figure(figsize=(14,6))
plt.plot(t_resp, yr)
plt.title("Respuesta yr(t) ante escalón unitario en E(s)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Salida yr(t)")
plt.grid()

plt.show()

# métricas de la respuesta
info = cs.step_info(Gatfm)
print("Tiempo de subida:", info['RiseTime'])
print("Tiempo de asentamiento:", info['SettlingTime'])
print("Valor en estado estacionario:", info['SteadyStateValue'])
print("Sobrepaso:", info['Overshoot'])
```

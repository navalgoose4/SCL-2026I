El **Lugar Geométrico de las Raíces (LGR)** es una herramienta gráfica fundamental en el análisis y diseño de sistemas de control lineales. Permite visualizar cómo varían las raíces del **polinomio característico** de un sistema en lazo cerrado cuando se modifica un parámetro del sistema, generalmente la **ganancia $ K $**.



## ⚙️ 1. Definición del LGR

Considérese un sistema de control con la siguiente **función de transferencia en lazo abierto**:

$$
G(s)H(s) = K \frac{N(s)}{D(s)} = K \frac{(s+z_1)(s+z_2)\cdots(s+z_m)}{(s+p_1)(s+p_2)\cdots(s+p_n)}
$$

El **polinomio característico en lazo cerrado** se obtiene de la ecuación de retroalimentación:

$$
1 + G(s)H(s) = 0 \quad \Rightarrow \quad D(s) + K N(s) = 0
$$

El **LGR** representa el conjunto de puntos en el plano complejo donde esta ecuación se satisface para **valores reales y positivos de $ K $**.



## 📐 2. Condición de ángulo y magnitud

De la ecuación de lazo cerrado:

$$
G(s)H(s) = -1
$$

obtenemos dos condiciones necesarias:

1. **Condición de magnitud:**
   $$
   |G(s)H(s)| = 1
   $$

2. **Condición de ángulo:**
   $$
   \angle G(s)H(s) = (2k+1)\pi, \quad k \in \mathbb{Z}
   $$

Estas condiciones determinan si un punto $ s = \sigma + j\omega $ pertenece o no al LGR.



## 📏 3. Regiones del eje real

Un punto sobre el eje real pertenece al LGR si el número de polos y ceros reales **a su derecha** es **impar**.  
Esto se deduce directamente de la condición de ángulo.



## 📊 4. Asintotas del LGR

Cuando $ K \to \infty $, las ramas del LGR tienden al infinito siguiendo líneas llamadas **asíntotas**, cuyo comportamiento depende de la diferencia entre el número de polos $ n $ y ceros $ m $:

$$
\text{Centroide: } \sigma_A = \frac{\sum p_i - \sum z_i}{n - m}
$$
$$
\text{Ángulos de las asíntotas: } \theta_A = \frac{(2k+1)\pi}{n - m}, \quad k = 0, 1, 2, \ldots
$$



## 🎯 5. Puntos de ruptura y ganancia

Los **puntos de ruptura o reunión** se encuentran resolviendo:

$$
\frac{dK}{ds} = 0, \quad \text{donde } K = -\frac{D(s)}{N(s)}.
$$

Además, la **ganancia** correspondiente a un punto $ s_0 $ sobre el LGR puede calcularse como:

$$
K = \frac{1}{|G(s_0)H(s_0)|}.
$$



## 🧭 6. Interpretación

- El LGR muestra cómo se mueven los **polos de lazo cerrado** al variar $ K $.
- Si las raíces se desplazan hacia el semiplano izquierdo, el sistema **gana estabilidad**.
- Si cruzan hacia el semiplano derecho, el sistema se **vuelve inestable**.



## 📈 7. Ejemplo básico

Para el sistema:

$$
G(s)H(s) = \frac{K}{s(s+2)(s+4)}
$$

El polinomio característico es:

$$
1 + \frac{K}{s(s+2)(s+4)} = 0 \quad \Rightarrow \quad s(s+2)(s+4) + K = 0
$$

Las raíces de esta ecuación, en función de $ K $, constituyen el **Lugar Geométrico de las Raíces**.

# 📘 Análisis del Lugar Geométrico de las Raíces (LGR) para un sistema de segundo orden con polo en el origen

Analicemos el siguiente sistema de control con retroalimentación unitaria:

$$
G(s)H(s) = \frac{\omega_n^2}{s(s + 2\zeta \omega_n)}
$$

donde:

- $ \zeta $: relación de amortiguamiento (adimensional)  
- $ \omega_n $: frecuencia natural no amortiguada (rad/s)



## ⚙️ Interpretación del sistema

Este sistema tiene dos polos:
$$
s_1 = 0, \qquad s_2 = -2\zeta \omega_n
$$
y **ningún cero**.  
El **LGR** describe cómo se mueven los polos de lazo cerrado cuando varía la ganancia $ K $.

El polinomio característico del sistema en lazo cerrado es:
$$
1 + G(s)H(s) = 0 \quad \Rightarrow \quad s(s + 2\zeta \omega_n) + K \omega_n^2 = 0
$$

De donde se obtiene la ecuación:
$$
s^2 + 2\zeta \omega_n s + K \omega_n^2 = 0
$$



## 📈 Interpretación geométrica

- Para $ K = 0 $, los polos de lazo cerrado están en $ s = 0 $ y $ s = -2\zeta \omega_n $.
- A medida que $ K $ aumenta, los polos se mueven a lo largo del **Lugar Geométrico de las Raíces**, que inicia en los polos de lazo abierto y termina en los ceros (reales o en infinito).
- Si $ \zeta < 1 $, el sistema exhibe **oscilaciones amortiguadas**; los polos se desplazan en el semiplano izquierdo formando una **curva parabólica** hacia arriba.
- Si $ \zeta > 1 $, el sistema es **sobreamortiguado**, y los polos permanecen en el eje real.



## 🧮 Forma general de las raíces

$$
s_{1,2} = -\zeta \omega_n \pm j \omega_n \sqrt{K - \zeta^2}
$$

El sistema es **estable** siempre que los polos permanezcan en el **semiplano izquierdo** ($\operatorname{Re}(s)<0$).

````python
# ==========================================================
# Lugar Geométrico de las Raíces para G(s)H(s) = w_n^2 / [s(s + 2ζw_n)]
# ==========================================================
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# Parámetros del sistema
wn = 1.0       # frecuencia natural (rad/s)
zeta = 0.3    # factor de amortiguamiento

# Función de transferencia G(s)H(s)
num = [wn**2]
den = [1, 2*zeta*wn, 0]  # s(s + 2ζwn)
G = ctrl.tf(num, den)

# Gráfico del LGR
plt.figure(figsize=(8,6))
rlist, klist = ctrl.rlocus(G, plot=True)
#rlist, klist = ctrl.rlocus(G, kvect=np.linspace(0,50,500), plot=True)
plt.title(f"Lugar Geométrico de las Raíces para ζ={zeta}, ωₙ={wn}")
plt.xlabel("Parte real (σ)")
plt.ylabel("Parte imaginaria (jω)")
plt.grid(True)
plt.show()

# Mostrar algunos polos para diferentes valores de K
for i, K in enumerate([0, 5, 10, 20]):
    poles = np.roots([1, 2*zeta*wn, K*wn**2])
    print(f"K={K:>4}: polos -> {poles[0]:.3f}, {poles[1]:.3f}")
````

````python
# ==========================================================
# Respuesta al escalón para distintos ζ y K
# Sistema de lazo abierto: G(s)H(s) = K * (wn^2) / [ s (s + 2 ζ wn) ]
# Lazo cerrado con realimentación unitaria: T(s) = K*G / (1 + K*G)
# ==========================================================
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl
from itertools import product

# Parámetros base
wn = 1.0                            # frecuencia natural (rad/s)
zeta_list = [0.1, 0.3, 0.7, 1.2]    # diferentes relaciones de amortiguamiento
K_list = [0.5, 1.0, 5.0, 10.0]      # ganancias a probar
t = np.linspace(0, 20, 2000)        # eje temporal común

def closed_loop_tf(wn, zeta, K):
    """Devuelve la FT de lazo cerrado T(s) para los parámetros dados."""
    # Lazo abierto sin K:  Go(s) = wn^2 / [ s (s + 2 ζ wn) ]
    Go = ctrl.tf([wn**2], [1, 2*zeta*wn, 0])   # s(s + 2ζwn)
    G = K * Go                                 # incluye K
    T = ctrl.feedback(G, 1)                    # realimentación unitaria
    return T

#  Figura 1: barrido de K para un ζ fijo (p.ej. zeta = 0.3)
zeta_fix = 0.3
plt.figure(figsize=(8,5))
for K in K_list:
    T = closed_loop_tf(wn, zeta_fix, K)
    t_out, y  = ctrl.step_response(T, T=t)
    plt.plot(t_out, y, label=f"K={K}")
plt.title(f"Respuesta al escalón: barrido de K (ζ={zeta_fix}, ωₙ={wn})")
plt.xlabel("Tiempo [s]")
plt.ylabel("Salida")
plt.grid(True)
plt.legend()
plt.show()

#  Figura 2: barrido de ζ para un K fijo (p.ej. K = 5.0)
K_fix = 5.0
plt.figure(figsize=(8,5))
for zeta in zeta_list:
    T = closed_loop_tf(wn, zeta, K_fix)
    t_out, y = ctrl.step_response(T, T=t)
    plt.plot(t_out, y, label=f"ζ={zeta}")
plt.title(f"Respuesta al escalón: barrido de ζ (K={K_fix}, ωₙ={wn})")
plt.xlabel("Tiempo [s]")
plt.ylabel("Salida")
plt.grid(True)
plt.legend()
plt.show()

#  Tabla de métricas para relacionar con el LGR
print("Métricas de respuesta (step_info)".center(70, "-"))
print(f"{'ζ':>5} {'K':>6} {'ts(2%)':>10} {'OS%':>8} {'tp':>8} {'wn_eq':>8}")
for zeta, K in product(zeta_list, K_list):
    T = closed_loop_tf(wn, zeta, K)
    info = ctrl.step_info(T)  # requiere 'control>=0.9'
    ts   = info.get('SettlingTime', np.nan)
    OS   = info.get('Overshoot', np.nan)
    tp   = info.get('PeakTime', np.nan)
    # frecuencia natural equivalente (segundo orden dominante si aplica)
    # a partir de polos dominantes:
    p = np.roots(T.den[0][0])  # coeficientes del denominador
    # tomamos los dos polos con parte real más grande (dominantes)
    p_sorted = sorted(p, key=lambda x: x.real, reverse=True)
    if len(p_sorted) >= 2:
        sigma = -p_sorted[0].real
        wd    = abs(p_sorted[0].imag)
        wn_eq = np.sqrt(sigma**2 + wd**2)
    else:
        wn_eq = np.nan
    print(f"{zeta:5.2f} {K:6.2f} {ts:10.3f} {OS:8.2f} {tp:8.3f} {wn_eq:8.3f}")

# Nota didáctica:
# - Al incrementar K, el LGR desplaza los polos a lo largo de sus ramas.
#   Observa cómo cambian OS (sobreimpulso), ts (tiempo de establecimiento) y tp.
# - Para ζ pequeño, el LGR cruza más cerca del eje imaginario ⇒ mayor oscilación.
# - Para ζ grande, los polos se mantienen más reales ⇒ respuestas más lentas y sin sobreimpulso.
````

````python
# ==========================================================
# LGR asociado a cada combinación (ζ, K)
# ==========================================================
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl
from itertools import product
from control import pzmap # Import pzmap

# ==========================================================
# LGR asociado a cada combinación (ζ, K)
# ==========================================================
plt.figure(figsize=(12,10))
plot_index = 1
total_plots = len(zeta_list) * len(K_list)

for zeta, K in product(zeta_list, K_list):
    Go = ctrl.tf([wn**2], [1, 2*zeta*wn, 0])   # Lazo abierto sin K
    G = K * Go                                 # incluye K
    T = ctrl.feedback(G, 1)                    # lazo cerrado

    plt.subplot(len(zeta_list), len(K_list), plot_index)
    # Graficar el LGR completo (sin marcar polos aún)
    rlist, klist = ctrl.rlocus(Go, kvect=np.linspace(0, 20, 500), plot=False)
    for r in rlist:
        plt.plot(np.real(r), np.imag(r), 'lightgray', lw=0.8)

    # Polos y ceros del sistema en lazo abierto
    # Use pzmap.pole and pzmap.zero
    plt.scatter(np.real(Go.poles()), np.imag(Go.poles()), c='r', marker='x', label='Polos L.A.')
    plt.scatter(np.real(Go.zeros()), np.imag(Go.zeros()), c='b', marker='o', facecolors='none', label='Ceros L.A.')

    # Polos de lazo cerrado para este K
    poles_T = ctrl.poles(T)
    plt.scatter(np.real(poles_T), np.imag(poles_T), c='k', marker='o', s=40, label='Polos L.C.')

    plt.title(f"ζ={zeta}, K={K}")
    plt.axvline(0, color='k', lw=0.5)
    plt.grid(True, alpha=0.3)
    if zeta == zeta_list[-1]:
        plt.xlabel("Parte real σ")
    if K == K_list[0]:
        plt.ylabel("Parte imaginaria jω")
    plot_index += 1

plt.suptitle("Lugar Geométrico de las Raíces (LGR) para cada combinación ζ y K", fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
````

# 🧩 Discusión de resultados

## 🔹 Análisis del Lugar Geométrico de las Raíces

El **LGR** obtenido muestra la trayectoria de los **polos del sistema en lazo cerrado** a medida que aumenta la ganancia $ K $.  
En este sistema de segundo orden con un polo en el origen:

$$
G(s)H(s) = \frac{\omega_n^2}{s(s + 2\zeta \omega_n)},
$$

los polos inician en $ s = 0 $ y $ s = -2\zeta \omega_n $ (puntos de lazo abierto).  
A medida que $ K $ crece, las raíces se desplazan desde estos polos hacia el **semiplano izquierdo**, describiendo una curva cuya forma depende de la **relación de amortiguamiento** $ \zeta $:

- Para $ \zeta < 1 $: los polos se mueven formando una **trayectoria compleja conjugada**, lo que produce oscilaciones amortiguadas.  
- Para $ \zeta = 1 $: el sistema es **críticamente amortiguado**, y los polos confluyen en un punto real doble.  
- Para $ \zeta > 1 $: los polos permanecen en el eje real, indicando un sistema **sobreamortiguado**.

El movimiento de los polos en el plano $ s $ está directamente relacionado con el comportamiento temporal del sistema:  
a medida que los polos se acercan al eje imaginario, el sistema se vuelve **más oscilatorio** y presenta un **mayor tiempo de establecimiento**.



## 🔹 Respuesta al escalón

En las simulaciones del lazo cerrado, se observa que:

- Un **menor valor de $ \zeta $** genera un **mayor sobreimpulso** y una **respuesta más oscilatoria**.  
- Al aumentar $ \zeta $**, el sistema responde más lentamente pero con menor sobreimpulso**, mostrando un comportamiento más estable y predecible.  
- Incrementar la **ganancia $ K $** acelera la respuesta (reduce el tiempo de establecimiento), pero también puede llevar el sistema a una condición inestable si los polos cruzan al semiplano derecho.

En términos generales:
$$
\text{Estabilidad} \Rightarrow \operatorname{Re}(s) < 0, \qquad
\text{Rapidez} \propto |\operatorname{Re}(s)|, \qquad
\text{Oscilación} \propto \operatorname{Im}(s).
$$



## 🔹 Conclusión

El análisis conjunto del **LGR** y de la **respuesta al escalón** permite comprender cómo los parámetros $ K $ y $ \zeta $ afectan la **estabilidad, rapidez y amortiguamiento** del sistema.  
Esta relación es esencial en el diseño de controladores, ya que el objetivo es seleccionar una ganancia $ K $ que logre un compromiso adecuado entre **velocidad de respuesta** y **estabilidad**, manteniendo los polos del sistema **suficientemente alejados del eje imaginario**.

## ⚙️ Modelo general de la planta

$$
G_p(s) = \frac{K_p \, e^{-L s}}{\tau s + 1}
$$

donde:
- $ K_p $: ganancia estática de la planta (°C/% de control)
- $ \tau $: constante de tiempo (s)
- $ L $: tiempo muerto o retardo (s)
- $ e^{-Ls} $: modela el retardo de transporte térmico

La ecuación de lazo cerrado para control proporcional ($ K_c $) es:

$$
1 + K_c G_p(s) = 0
$$

El **LGR** describe cómo se mueven los polos de lazo cerrado en el plano complejo cuando varía la ganancia $ K_c $.

````python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# ---------- Parámetros de la planta (EDITE AQUÍ) ----------
Kp  = 2.0     # ganancia de la planta en lazo abierto
tau = 30.0    # constante de tiempo [s]
L   = 3.5     # tiempo muerto [s]

# Orden de Padé para aproximar el retardo (1 o 2 recomendado)
pade_order = 1

# Conjunto de ganancias Kc para responder al escalón (lazo cerrado) "pueden variar estos valores"
Kc_list = [0.2, 0.5, 1.0, 2.0]

# Rango de K para el LGR
K_rlocus = np.linspace(0, 10, 800)

# ---------- Construcción de la planta con Padé ----------
# Planta sin retardo: Kp/(tau*s + 1)
s = ctrl.TransferFunction.s
Gp_nodelay = Kp / (tau*s + 1)

# Aproximación de Padé del retardo e^{-Ls}
num_p, den_p = ctrl.pade(L, pade_order) if L > 0 else ([1.0], [1.0])
delay_pade = ctrl.tf(num_p, den_p)

# Planta aproximada racional
#Planta lazo abierto con tiempo muerto
Gp = Gp_nodelay * delay_pade

print("Gp(s) racional (con Padé):")
print(Gp)
````

Expresando el sistema en lazo abierto:
   $$
   G(s)H(s) = K_c G_p(s)
   $$
   
   Se asume realimentación unitaria, es decir $H(s)=1$. Y como $K_c$ lo varía rlocus:

   $$G(s)=G_p(s)$$

````python
# ==========================================================
# Lugar Geométrico de las Raíces para Gp(s)
# ==========================================================
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# Gráfico del LGR
plt.figure(figsize=(8,6))
rlist, klist = ctrl.rlocus(Gp, plot=True)
#rlist, klist = ctrl.rlocus(G, kvect=np.linspace(0,50,500), plot=True)
plt.title(f"Lugar Geométrico de las Raíces variando Kc")
plt.xlabel("Parte real (σ)")
plt.ylabel("Parte imaginaria (jω)")
plt.grid(True)
plt.show()

# Mostrar algunos polos para diferentes valores de Kc
poles = Gp.poles()
for pol in poles:
  print(f"polos -> {pol:.3f}")
````

Análisis del tiempo muerto $L$

````python
L_values = [0, 2, 5, 10]

plt.figure(figsize=(10,8))

for L in L_values:
    # Padé
    num_p, den_p = ctrl.pade(L, 1) if L > 0 else ([1], [1])
    delay = ctrl.tf(num_p, den_p)

    # Planta
    Gp = (Kp / (tau*s + 1)) * delay

    # LGR
    rlist, klist = ctrl.rlocus(Gp, kvect=K_rlocus, plot=False)
    
    # Graficar solo polos (parte real vs imaginaria)
    plt.plot(np.real(rlist), np.imag(rlist), label=f"L = {L}")

plt.axvline(0)  # eje imaginario
plt.title("Efecto del tiempo muerto en el LGR")
plt.xlabel("Parte real")
plt.ylabel("Parte imaginaria")
plt.legend()
plt.grid()
plt.show()
````

Al aumentar el tiempo muerto L, las trayectorias del lugar geométrico de las raíces se desplazan hacia la derecha en el plano complejo. Esto provoca que los polos se acerquen al eje imaginario, reduciendo la estabilidad del sistema. Para valores grandes de L, el sistema puede volverse inestable con menores valores de ganancia.

A medida que aumenta el tiempo muerto, los polos del sistema tienden a formar pares complejos conjugados, lo que indica una respuesta más oscilatoria. Esto implica mayor sobreimpulso y menor amortiguamiento.

El incremento del tiempo muerto reduce el margen de fase debido al desfase adicional introducido en el sistema. Como consecuencia, el margen de ganancia también disminuye, haciendo que el sistema sea más sensible a cambios en la ganancia y más propenso a la inestabilidad.

    El tiempo muerto introduce un retardo puro que no puede ser compensado directamente por un controlador proporcional, ya que este no anticipa la dinámica futura del sistema. Para mitigar este efecto, se pueden emplear estrategias más avanzadas:

    - **Control PID:** permite mejorar el desempeño dinámico al incluir acción integral (elimina error en estado estacionario) y derivativa (anticipa cambios), aunque su efectividad se ve limitada cuando el tiempo muerto es significativo.
    - **Predictor de Smith:** es una técnica especialmente diseñada para sistemas con tiempo muerto. Utiliza un modelo interno del proceso para predecir la salida futura, compensando el retardo y mejorando la estabilidad y la respuesta transitoria.

    En general, mientras mayor sea el tiempo muerto, más necesario es utilizar estrategias de control que incorporen predicción, ya que el retardo reduce significativamente los márgenes de estabilidad.
